```python

# one hot encoder
# it is a technique used to
# convert text data into numerical data by creating seperate columns for each
# category and filling thems with 0s and 1s
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
df = sns.load_dataset("tips")
```

```python
df
```

```python
sex_data = df[["sex"]] # Pass a DataFrame with a single column
encoder = OneHotEncoder(sparse_output = False)
encoded = encoder.fit_transform(sex_data)
print(encoded)
```

```python
print(encoder.get_feature_names_out())
```

```python
import pandas as pd
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns = encoder.get_feature_names_out()
)
encoded_df
```

```python
d = df.drop("sex", axis = 1)
```

```python
d
```

```python
final_df = pd.concat([d,encoded_df], axis = 1)
final_df
```

```python
# Task
import pandas as pd
import seaborn as sns
df = sns.load_dataset("diamonds")
print(df.head(5))
print(df.shape)
print(df.columns)
```

```python
print(encoder.get_feature_names_out())
```

```python
df
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns = encoder.get_feature_names_out()
)
encoded_df
```

```python
d = df.drop("cut", axis = 1)
final_df = pd.concat([d,encoded_df], axis = 1)
final_df
display(final_df.head(5))
```

```python
color_data = df[["color"]]
encoder = OneHotEncoder(sparse_output = False)
encoded = encoder.fit_transform(color_data)
print(encoded)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
cut_data = df[["cut"]]
encoder_cut = OneHotEncoder(sparse_output=False)
encoded_cut = encoder_cut.fit_transform(cut_data)
encoded_df_cut = pd.DataFrame(encoded_cut, columns=encoder_cut.get_feature_names_out())
color_data = df[["color"]]
encoder_color = OneHotEncoder(sparse_output=False)
encoded_color = encoder_color.fit_transform(color_data)
encoded_df_color = pd.DataFrame(encoded_color, columns=encoder_color.get_feature_names_out())
clarity_data = df[["clarity"]]
encoder_clarity = OneHotEncoder(sparse_output=False)
encoded_clarity = encoder_clarity.fit_transform(clarity_data)
encoded_df_clarity = pd.DataFrame(encoded_clarity, columns=encoder_clarity.get_feature_names_out())
df_processed = df.drop(columns=["cut", "color", "clarity"])
final_df = pd.concat([df_processed, encoded_df_cut, encoded_df_color, encoded_df_clarity], axis=1)
display(final_df.head(5))
```
