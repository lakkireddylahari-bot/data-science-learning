```python
#groupby enables categorywise  analsysis instead of row by row anslaysis
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/weather_by_cities.csv")
```

```python
df
```

```python
#find maximum temperature in each of the cities
#find average wind speed per city
```

```python
g = df.groupby("city")
print(g)
```

```python
for city,city_df in g:
  print(city)
  print(city_df)
```

```python
g.get_group('mumbai')
```

```python
df.columns
```

```python
df.groupby("city")["temperature"].max()
```

```python
# find average wind speed per city
df.groupby("city")["windspeed"].mean()
```

```python
#find the total wind speed for each city
df.groupby("city")["windspeed"].sum()
#find the highest  wind speed for each city
```

```python
#find the highest  wind speed for each city
df.groupby("city")["windspeed"].max()
```
