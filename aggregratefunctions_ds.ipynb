```python
# aggregrate functions are functions used to summarize
# multiple data values into a single meaningful result
# sum,max,min,mean
```

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/drive/MyDrive/datsetsfordatasciencebytexl/student_dataset_20rows.xlsx")
```

```python
df.head()
```

```python
print(df["Marks"].sum()) #sum : it will calucualtes sum all the data in that column
```

```python
print(df["Marks"].mean()) #it will calcualtes the mean
```

```python
print(df["Marks"].median()) #it will calcualtes the median
```

```python
print(df["Marks"].min()) #it will calculates the min
```

```python
print(df["Marks"].max()) #maximum value
```

```python
print(df["Marks"].count())
```

```python
print(df["Marks"].std())
```

```python
print(df["Marks"].var())
```

```python
print(df["Marks"].describe())
```

```python
print(df["Marks"].agg(["sum", "mean"]))
```
