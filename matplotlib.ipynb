```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40,50,60])
plt.show()
# x = [0,1,2,3,4,5] # horizontal
# y = [0,2,4,6,8,10] # vertical
# plt.plot(x,y)
```

```python
x = [1,2,3,4,5,6,7]
y = [5,20,60,50,50,50,100]
plt.plot(x,y)
plt.show()
```

```python
days = [1,2,3,4,5]
sales = [100,200,5,500,700]
plt.figure(figsize = (5,7))
# by using figure we can draw our graph in particular weight of height
plt.plot(days,sales)
plt.title("Sales Per Day")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
# task
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
temperatures = [30,32,35,34,36]
plt.figure(figsize = (8,4))
plt.plot(days, temperatures)
plt.title("Temperature Analysis")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
days = [1,2,3,4,5]
productasales = [100,200,200,300,400]
productbsales = [200,300,400,500,600]
plt.plot(days,productasales,label = "Product A")
plt.plot(days,productbsales,label = "Product B")
plt.title("Sales Analysis")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
