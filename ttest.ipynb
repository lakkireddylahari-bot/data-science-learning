```python
from scipy.stats import ttest_1samp
import numpy as np
marks =[49,48,49,49]
# H0 = 50
t_statistic, p_value = ttest_1samp(marks,50)
print(t_statistic)
print(np.mean(marks))
if p_value <= 0.05:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

```python
from scipy.stats import ttest_ind
import numpy as np
old_method_marks = [70,75,72,68,74,71,69,73]
new_method_marks = [78,80,82,77,79,81,83,80]
# Ho old_mean = new_mean
print("old method mean", np.mean(old_method_marks))
print("new method mean", np.mean(new_method_marks))
t_stat, p_value = ttest_ind(old_method_marks, new_method_marks)
print("tstat", t_stat)
print("p_value", p_value)
if p_value <= 0.05:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```
