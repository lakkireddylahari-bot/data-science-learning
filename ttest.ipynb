```python
from scipy.stats import ttest_1samp
import numpy as np
marks =[49,48,49,49]
# H0 = 50
t_statistic, p_value = ttest_1samp(marks,50)
print(t_statistic)
print(np.mean(marks))
if p_value <= 0.05:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```
