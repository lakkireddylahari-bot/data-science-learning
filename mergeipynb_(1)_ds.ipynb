```python
#merge
# in real world  data analysis data is usually stored in multiple tablles
# ex:
# studnet deatils in one table
# student marks in another table
# so to combine these tables we will use merge
# pd.merge(
#     dataset1,
#     dataset2,
#     on =column name
#     how = type
# )
```

```python
import pandas as pd
```

```python
student_df = pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/students.csv")
print(student_df)
```

```python
marks_df = pd.read_csv("/content/drive/MyDrive/datsetsfordatasciencebytexl/marks.csv")
print(marks_df)
```

```python
# inner merge
# it will takes only common data from both tables
result = pd.merge(
    student_df ,
    marks_df,
    on = 'student_id',
    how = 'inner'
)
print(result)
```

```python
result_l = pd.merge(
    student_df,
    marks_df,
    on ='student_id',
    how ='left'

)
print(result_l)
```

```python
result_l = pd.merge(
    student_df,
    marks_df,
    on ='student_id',
    how ='right'

)
print(result_l)
```

```python
result_l = pd.merge(
    student_df,
    marks_df,
    on ='student_id',
    how ='outer'

)
print(result_l)
```
